```python
import torch
from pathlib import Path
import os
import cv2

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# Image and output paths
image_folder = '../Data/medical_photos/'
output_folder = '../detection_output/'
os.makedirs(output_folder, exist_ok=True)

# Process images
for image_path in Path(image_folder).glob('*.*'):
    img = str(image_path)
    print(f"Processing image: {img}")
    
    results = model(img)
    results.render()
    
    output_image_path = os.path.join(output_folder, os.path.basename(img))
    
    for img_result in results.ims:  
        # Convert from RGB to BGR for OpenCV saving
        img_result_bgr = cv2.cvtColor(img_result, cv2.COLOR_RGB2BGR)
        cv2.imwrite(output_image_path, img_result_bgr)

    # Log if no objects were detected
    if len(results.pandas().xyxy[0]) == 0:
        print(f"No objects detected in {img}")
```

```python
df_results = results.pandas().xyxy[0]  # DataFrame with detection results
df_results 
```

```python
import psycopg2

def store_detection_results(df_results):
    connection = psycopg2.connect(
        database="m", 
        user="your_db_user", 
        password="your_db_password", 
        host="localhost", 
        port="5432"
    )
    cursor = connection.cursor()

    for index, row in df_results.iterrows():
        query = """
        INSERT INTO detection_results (image, label, confidence, xmin, ymin, xmax, ymax)
        VALUES (%s, %s, %s, %s, %s, %s, %s);
        """
        cursor.execute(query, (
            row['image'], row['name'], row['confidence'], 
            row['xmin'], row['ymin'], row['xmax'], row['ymax']
        ))

    connection.commit()
    cursor.close()
    connection.close()

# Call the function after extracting detection results
store_detection_results(df_results)
```
